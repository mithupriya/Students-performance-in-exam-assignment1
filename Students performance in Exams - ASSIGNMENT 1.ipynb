```python
import pandas as pd  # data processing ,CSV file
import numpy as np   #linear algebra
import matplotlib.pyplot as plt  #visualisation
data=pd.read_csv('/home/nbuser/library/StudentsPerformance.csv')
data.head
```

```python
# checking for nulls
data.isnull().sum()
```

```python
# Adding a column of Total score in dataframe
data['total score'] = data['math score']  +data ['reading score'] + data['writing score']
data.head()
```

```python
data['total score'].plot(kind='box')
plt.show()
```

```python
# students who have failed in all three subjects(less than 40)
data_fail_all= data[(data['math score'] < 40) & (data['reading score'] < 40) & (data['writing score'])]
data_fail_all.head()
```

```python
# Toppers
data_topper = data[(data['math score'] > 90) & (data['reading score'] > 90) & (data['writing score'] > 90)]
data_topper.head()
```

```python
data_topper.sort_values('total score',ascending=False).head()
```

```python
# number of students against scores in all three subjects
plt.hist([data['math score'], data['reading score'], data['writing score'] ], color=['red','yellow','blue'])
plt.title('Number of students against scores')
plt.xlabel('score')
plt.ylabel('number of students')
plt.legend(['math','reading','writing'])
plt.show()
```

```python
# ( From the above histogram it is clear that Number of students with higher score in mathematics have dropped below than reading and 
#  writing which may show that is easier to get a higher score in reading and writing than mathematics)
```

```python
# comparison of total score and gender
data.boxplot('total score' , by='gender')
plt.show()
```

```python
# comparison of total score and lunch
data.boxplot('total score' , by='lunch')
plt.show()
```

```python
# comparison of total score and test preparation course
data.boxplot('total score' , by='test preparation course')
plt.show()
```

```python
# From the above boxplot it clears that the students who have completed the test preparation course have greater total score than students 
# who have not completed the test preparation course.
```
